```python
import os

def make_salt_hex(length: int) -> str:
    """
    Генерирует строку случайных байт указанной длины в hex-формате.

    Args:
        length: Длина соли в байтах

    Returns:
        Строка в hex-формате длиной length * 2 символов
    """
    # Генерируем случайные байты с помощью os.urandom
    random_bytes = os.urandom(length)

    # Преобразуем в hex-строку
    return random_bytes.hex()
```

```python
# Генерация соли длиной 8 байт (16 символов hex)
salt = make_salt_hex(8)
print(salt)  # 'a7f3c91e2b4d8f0a'

# Генерация соли длиной 16 байт (32 символа hex)
salt = make_salt_hex(16)
print(salt)  # '9c3f1a7e2b4d8f0a5c6b7d8e9f0a1b2c'

# Для хеширования паролей обычно используют соль 16-32 байта
password_salt = make_salt_hex(16)
```
